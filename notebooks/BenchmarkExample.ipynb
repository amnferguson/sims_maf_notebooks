```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Import the sims_maf modules needed.

```python
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.benchmarks as benchmarks
import lsst.sims.maf.db as db
```

'Benchmarks' combine a particular metric, a slicer, and a sqlconstraint. You can also add additional provenance/metadata information to a given benchmark, in the form of the (opsim) runName and a metadata comment. This information will be used to generate outfile filenames and plot labels. A given benchmark will also store the metric values calculated at each point of the slicer (after this is generated via the 'benchmarkGroup' methods). 

To instantiate a benchmark, first instantiate a metric and slicer, then the benchmark to store the combination. Here we'll calculate the minimum airmass value, at each point in a healpix grid.

```python
metric = metrics.MinMetric('airmass')
slicer = slicers.HealpixSlicer(nside=64)
sqlconstraint= 'filter = "r"' 
minairmass = benchmarks.Benchmark(metric, slicer, sqlconstraint)
```

Here is another example of a benchmark, this time reusing the same slicer, but changing the metric - this time we'll count the number of visits at each point in the healpix grid.

```python
metric = metrics.CountMetric('expMJD')
nvisits = benchmarks.Benchmark(metric, slicer, sqlconstraint)
```

We bundle benchmarks with the same sql constraint together into a dictionary, which will be passed to the benchmarkGroup to run the benchmarks, calculating the metric values.

```python
bDict = {'min airmass hp':minairmass, 'nvisits hp':nvisits}
```

Now we set up our connections to the opsim database, and a 'results database' where we will store information about the benchmarks which were run.

```python
opsdb = db.OpsimDatabase('sqlite:///enigma_1189_sqlite.db')
outDir = 'test'
resultsDb = db.ResultsDb(outDir=outDir)
```

Ready to do some calculating! Instantiate the BenchmarkGroup by passing it the dictionary of benchmarks we want to run (they must have the same sql constraint), the opsim database object, the output directory and the results database object.

```python
bgroup = benchmarks.BenchmarkGroup(bDict, opsdb, outDir=outDir, resultsDb=resultsDb)
```

Query the data from the opsim database. Note that MAF determines what columns you need from the database for your metrics and slicers, and only queries for those.

```python
bgroup.getData()
```

Run all of the benchmarks in your benchmark dictionary, calculating the metric values at each slice point. MAF does this in an efficient manner, caching results where possible and iterating through the slicer once for all related metrics.

```python
bgroup.runAll()
```

The results of the metric calculation are stored in the benchmark, as 'metricValues' - a numpy masked array.

```python
print nvisits.metricValues
```

We can now save the benchmark results to disk. 

```python
bgroup.writeAll()
```
